# Python Data Type(자료 타입)

*   숫자 타입
    *   정수(int)
    *   실수(float)
*   문자열 타입(str)
*   논리 타입(bool): True/False

## 숫자타입

```python
# 변수 선언, 값 할당: 변수_이름 = 값
n = 123
print(type(n))    # type(변수_이름): 변수의 타입을 리턴하는 함수
print('n =', n)
```

```python
n = 3.14  # 변수의 저장된 값을 변경.
print(type(n))
print('n =', n)
```

## 문자열 타입

파이썬에서 문자열은 작은 따옴표와 큰 따옴표를 모두 사용할 수 있다.

```python
name = '오쌤'
print(f'안녕하세요, {name}!')
print(type(name))
```

```python
b = True # 파이썬의 키워드, 변수 이름, 함수 이름 등은 대소문자를 구분한다.
print(type(b))
print('b=' , b)
```

```python
b = 5 < 0
print(b)
```

## 명시적 타입 변환
*   `int(arg)`: 아규먼트로 전달된 값을 정수로 변환해서 리턴하는 함수.
    *   문자열 -> 정수
    *   실수 -> 정수

*   `float(arg)`: 아규먼트를 실수로 변환해서 리턴하는 함수.
    *   문자열 -> 실수
    *   정수 -> 실수
    
*   `str(arg)`: 아규먼트를 문자열로 변환해서 리턴하는 함수.
    *   정수, 실수, ... -> 문자열

```python
x = input('숫자를 입력하세요: ')
print('x =', x)
print(type(x))

x = int(x);
print(type(x))
```

```python
x = float(input('숫자 입력 : '))
print('x =',x)
print(type(x))
```

# 연산자(Operator)

## 할당(assignment) 연산자

*   '=' : 연산자 오른쪽의 값을 연산자의 왼쪽의 변수에 할당(저장).


## 산술(arithmetic) 연산자
*   종류: ' + , - , * , / , **(거듭제곱) , //(몫) , %(나머지) '
*   산술 연산자는 일반적으로 숫자 타입(int, float)의 값들에서 사용.
*   일부 산술 연산자는 문자열과 같은 타입에서도 사용 가능.

```python
x, y = 10, 3
print(f'{x} + {y} = {x+y}')
print(f'{x} - {y} = {x-y}')
print(f'{x} * {y} = {x*y}')
print(f'{x} / {y} = {x/y}')
print(f'{x}의 {y} 거듭제곱 = {x**y}') # x의 y의 거듭제곱
print(f'{x}를 {y}로 나눈 몫 = {x//y}')
print(f'{x}를 {y}로 나눈 나머지 = {x%y}')
```

문자열 (str) 타입에서 사용하는 산술 연산자:

*   `문자열 + 문자열` :  concatnate. 문자열 이어 붙이기.
*   `문자열 * 정수` : 문자열 정수만큼 복사.

```python
print('hello' + 'python')
print('hello' * 3)
```

## 비교 연산자, 논리 연산자

*   비교 연산자: `==, !=, > , >= , < , <=`
*   논리 연산자: `and, or, not`


```python
x, y = 100, 200
print(x == y)
print(x != y)
print(x > y)
print(x >= y)
print(x < y)
print(x <= y)
```

```python
print((x > 50) and (x< 150))
print((x < 50) or (x > 150))
```

## 복합 할당 연상잔

*   종류: ` +=, -=, *=, /=, **=, //=, %=, ...`
*   산술 연산자와 할당 연산자를 합쳐서 만들어진 연산자.

```python
x = 10
print (x)
x += 1  # x = x + 1과 같다. 설명)(1) x + 1 연산 수행 (2) 그 결과를 변수 x에 저장
print(x)
```
