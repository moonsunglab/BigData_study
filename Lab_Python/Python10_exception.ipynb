# try-except 구문

```
try:
    (1) 실행할 코드 블록
except [에러이름 [as 별칭]]:
    (2) 예외가 발생했을 때 예외를 처리하는 코드 블록
[else:
    (3) try 블록의 모든 코드가 예외없이 정상 수행됐을 때 실행할 코드 블록]
[finally:
    (4) 예외 발생 여부에 상관없이 항상 실행할 코드 블록]
```

*   try-except 구문의 실행 순서
    *   예외가 발생하는 경우: (1) -> (2) -> (4)
    *   예외가 발생하지 않는 경우: (1) -> (3) -> (4)
*   except 절에서 에러 이름을 생략하면 모든 종류의 에러를 처리할 수 있음.
*   except 절에서 에러 이름을 지정하면, 선언된 에러만 처리할 수 있음.
*   처리해야 할 에러의 종류가 여러 개인 경우 except 절을 여러 개 사용할 수도 있음.


````python
try:
    int1 = int(input('int1: '))
    int2 = int(input('int2: '))

    print(f'{int1} + {int2} = {int1 + int2}')
    print(f'{int1} - {int2} = {int1 - int2}')
    print(f'{int1} x {int2} = {int1 * int2}')
    print(f'{int1} / {int2} = {int1 / int2}')

except ValueError:
    print('int1 & int2 는 정수로 입력해야 합니다.')

except ZeroDivisionError:
    print('0으로는 나눌 수 없습니다.')

else:
    print('사칙연산을 정상 수행')

finally:
    print('코드 셀 종료.')
````

````python
try:
    int1 = int(input('int1: '))
    int2 = int(input('int2: '))

    print(f'{int1} + {int2} = {int1 + int2}')
    print(f'{int1} - {int2} = {int1 - int2}')
    print(f'{int1} x {int2} = {int1 * int2}')
    print(f'{int1} / {int2} = {int1 / int2}')

except:
    print('오류가 발생해서 코드 실행을 멈춥니다.')
````

````python
import random
````

````python
numbers = [random.randrange(100) for _ in range(10)]
numbers
````

````python
try:
    index = int(input('인덱스를 입력: '))
    print(f'numbers[{index}] = {numbers[index]}')

except IndexError:
    print('인덱스는 0부터 9까지 번호를 입력해야합니다. ')

except ValueError:
    print('인덱스는 정수여야합니다.')

else:
    print('출력완료')
finally:
    print('프롬프트 종료')
````

````python
# 오쌤
try:
    index = int(input('인덱스를 입력: '))
    print(f'numbers[{index}] = {numbers[index]}')
except ValueError as e:
    print(e)        # ValueError의 오류메세지가 출력

except IndexError as e:
    print(e)        #IndexError
````

사용자에게 숫자(실수float)을 입력 하라고 프롬프트를 보여주고,
*   사용자가 입력한 문자열이 숫자로 변환될 수 있는 경우라면, 그 숫자를 출력하고 셀을 종료
*   사용자가 입력한 문자열이 숫자로 변환될 수 없는 경우라면, 그숫자 입력 프롬프트를 보여주고 입력 과정을 반복

````python


while True :
    try:
        index = float(input('숫자입력: '))
        print(index)

    except:
        print('사용자가 입력한 문자열이 숫자로 변환될 수 없습니다.')
    else:
        print('출력완료')
        break
    finally:
        print('프롬프트 종료')



````
