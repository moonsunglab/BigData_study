# for loop(반복문)

```
for 변수 in iterable=type:
    반복해서 실행할 코드 블록
```

*   iterable-type: 반복(iteration)이 가능한 데이터 타입.
    *   str(문자열), range, list, tuple, dict, ...
*   for 반복문은 iterable-type에서 원소(아이템)들을 순서대로 하나씩 꺼내서 변수에 저장하고, 코드 블록을 실행.
*   for 반복문은 iterable-type에서 더 이상 꺼낼 원소가 없을 때 종료됨.

````python
for s in '안녕하세요':
    print(s)
````

# for 반복문과 `range()`함수

````python
# range(start, stop): start가 stop보다 작은 경우, [start, end)범위의 1씩 증가하는 정수들을 만들어 줌.
for x in range(1, 6):
    print(x)
````

````python
# range(stop): [0, stop) 범위의 1씩 증가하는 정수들을 만들어 줌.
for x in range(5):
    print(x)
````

````python
# range(start, stop, step): start가 stop보다 작고, step은 양수인 경우,
# start <= x < stop 범위의 step만큼씩 증가하는 정수들을 만들어 줌.
for x in range(0, 10, 2):
    print(x)
````

````python
for x in range(1,10,2):
    print(x)
````

````python
# range(start, stop, step): start가 stop보다 크고 step이 음의 정수인 경우,
# start >= x > stop 범위의 step만큼씩 감소하는정수들을 만들어줌
for x in range(5, 0, -1):
    print(x)
````

````python
for x in range(1, 6):
    print('*' * x)
````

````python
for x in range (5, 0 ,-1):
    print('*' * x)
````

# `break`와 `continue`

*   `break`: iteration을 종료.
*   `continue`: 그다음 iteration 과정을 수행

````python
for x in range(5):
    if x == 2:
        break
    print(x)
````

````python
for x in range(5):
    if x == 2:
        continue
    print(x)
````

for-in 반복문을 사용해서 구구단 2단은 2x2까지.. 9단은 9x9까지 출력코드 생성

````python
# break 사용
for x in range(2,10):
    print(f'=== {x}단 ===')
    for y in range(1,10):
        print(f'{x} X {y} = {x *y}')
        if x == y :
            break
    print(f'----------------')
````

````python
# break 사용x
for x in range(2,10):
    print(f'=== {x}단 ===')
    for y in range(1,x + 1):
        print(f'{x} X {y} = {x *y}')
    print(f'----------------')
````

# `print()`함수

*   탈출 문자열(escape string)
    *   '\t': 탭(tab)
    *   '\n': 줄바꿈(new line)
*   `print()`의 파라미터(parameter)
    *   `end`: print 함수가 아규먼트들을 콘솔에 출력한 후 마지막에 어떤 문자열을 추가할 것인지를 설정. 기본값은 '\n'(줄바꿈)
    *   `sep`: seperator(구분자). print 함수의 아규먼트들을 출력할때, 각각의 아규먼트들을 구분하기위한 문자(열)을 설정. 기본값은 공백(space)

````python
print('안녕 hello bon jour')
print('안녕\thello\tbon jour')
print('안녕\nhello\nbon jour')
````

````python
print('안녕하세요', end= '...')
print('hello')
print('bon jour')
````

````python
print('hello','python','파이썬', sep = ',')
````

## EX 1. 아래와 같이 출력하는 코드를 작성하세요.
```
1 little 2 little 3 little indian
4 little 5 little 6 little indian
7 little 8 little 9 little indian
10 little indian boys!
```

````python
for x in range(1, 11, 3):
    y = 'little'
    if x == 10:
        print(f'{x} {y} indian boys!')
        break
    print (f'{x} {y} {x + 1} {y} {x + 2} {y} indian')
````

````python
# 오쌤 풀이 for in
for n in range(1, 11):
    print(f'{n} little', end = ' ')
    if n % 3 ==0:        #요거 히트다
        print('indian')

print('indian boys!')
````

````python
# 오쌤풀이 while
n = 1
while n < 11:
    print(f'{n} little', end= ' ')
    if n % 3 == 0:
        print ('indian')
    n += 1
print('indian boys!')
````

## Ex 2. 아래와 같이 출력하는 코드를 작성하세요.
```
1   2   *   4   5   *   7   8   *   10
11  12  *   ... 100
```

````python
for x in range(1, 101):

    if '0' in str(x):
        if '3' in str(x):
            print('*')
        elif '6' in str(x):
            print('*')
        elif '9' in str(x):
            print('*')
        else:
            print(x)
    elif '3' in str(x):
        print('*', end= '  ')
    elif '6' in str(x):
        print ('*', end = '  ')
    elif '9' in str(x):
        print ('*', end = '  ')
    elif '0' in str(x):
        print(x)
    else:
        print(x, end= '  ' )
````

````python
# 오썜풀이
for n in range(1,101):
    n_1 = n % 10    # 10으로 나눈 나머지 = 숫자 n의 1의 자릿수
    n_10 = n // 10  # 10으로 나눈 몫 = 숫자 n의 10의 자릿수
    b_1 = (n_1 == 3) or (n_1 == 6) or (n_1 == 9)         # 1의 자릿수가 3 또는 6 또는 9 이면
    b_10 =  (n_10 == 3) or  (n_10 == 6) or (n_10 == 9)   # 10의 자릿수가 3 또는 6 또는 9 이면

    if b_1 and b_10:
        print('**', end = '\t')
    elif b_1 or b_10:
        print('*', end = '\t')
    else:
        print(n, end = '\t')

    if n % 10 == 0:
        print()
````

````python
x = 1
print(x in (1,2,3))
print(x in (3, 6, 9))
# 변수(식) in (값1, 값2, ...): 변수(식)이 값들 중에 하나면 True. 그렇지 않으면 False.
````

````python
for n in range(1, 101):

    n_1 = n % 10
    n_10 = n // 10

    if n_1 in (3, 6, 9) and n_10 in (3, 6, 9):
        print('**', end = '\t')
    elif n_1 in (3, 6, 9) or n_10 in (3, 6, 9):
        print('*', end = '\t')
    else:
        print(n, end = '\t')

    if n % 10 == 0:
        print()
````
