# `str`클래스 메서드

```python
s = 'heLLo, pyTHon! hello PYTHON! 안녕하세요...'
```

```python
print(s)        # print 함수는 객체(s)의 __str__ 메서드가 리턴하는 문자열을 콘솔에 출력
```

```python
s               # 노트북의 코드셀 출력(expression)은 __repr__ 메서드가 리턴하는 문자열을 콘솔에 출력
```

```python
result = s.upper()  # str.upper(): 대문자로 변환한 "새로운" 문자열을 리턴.
print(result)
print(s)    # 원본 문자열 s는 변경되지 않음!
```

```python
s.lower()
```

```python
s.capitalize()  # 문자열의 첫번째 글자만 대문자, 나머지는 소문자로 변환한 문자열을 리턴.
```

```python
print(s)
print(s.count('l'))
print(s.count('LL'))
```

```python
# 대/소문자 구분 없이 'py' 문자열이 몇 번 등장하는지
# 내 방법
result = s.lower()
print(result)
print(result.count('py'))
# 오쌤 방법
s.lower().count('py')
```

```python
email = 'gildong@itwill.com'

# 문자열 email에서 아이디(gildong)와 도메인(itwill.com)을 구분해서 출력

# 내방법
print(email[:email.find('@')])
print(email[email.find('@'):])

# 오쌤방법
# 문자열 email에서 '@' 문자열이 등장하는 인덱스
idx = email.find('@')
print(idx)
print(email[:idx])
print(email[idx + 1 :])
# 다른점: idx라는 변수를 정해줌. +1이 힛또다.
```

```python
msg = '파이썬 참 쉽죠? 파이썬 열심히 공부합시다.'
result = msg.split()    # str.split(): 문자열 str을 공백(space, 줄바꿈, 탭)으로 구분한 문자열들의 리스트를 리턴.
result
```

```python
csv_str = 'hello,python,안녕,오쌤,12345'
csv_str.replace(',','::')
# str.replace(old,new): 문자열 str에서 등장하는 모든 old 문자열을 new 문자열로 대체한 문자열을 리턴.
```

```python
msg = '안녕? 안녕하세요! Hello, Python.'
# 문자열 msg에서 문장부호(?,!,.)들을 제거한 단어들로만 이루어진 리스트
# 내방법
msg = msg.replace('!', '')
msg = msg.replace('?', '')
msg = msg.replace(',', '')
msg = msg.replace('.', '')
print(msg.split())

#오쌤방법
msg = '안녕? 안녕하세요! Hello, Python.'
msg.replace('!', '').replace('?', '').replace(',', '').replace('.', '').split()

# 리턴값이 문자열이라서 다시 .을 써서 메소드를 호출할 수 있다.

# 정규표현식 찾아보기.
```

```python
sentence = '               hello, Python!         안녕하세요.               '
print(sentence)
print(sentence.lstrip())            #left에 있는 공백을 strip한다.
print(sentence.rstrip())            #right에 있는 공백을 strip한다.
print(sentence.strip())
```
