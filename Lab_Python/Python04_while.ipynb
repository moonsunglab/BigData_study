# while 반복문(loop)

```
[초기화 문장]
while 조건식:
    조건을 만족하는 동안에 반복할 코드 블록
    [조건을 변경할 수 있는 문장]
```

* 조건식을 만족하는 동안에 블록 내부를 실행.
* 조건식을 만족하지 못하면 (조건식이 False가 되면) 반복을 멈춤.
* `while 조건식` 다음에는 반드시 콜론(:)을 사용해야함.
* 블록 안에서 들여쓰기는 일정한 크기로 작성.



````python
# 1 ~ 5까지 순서대로 출력
n = 1
while n <= 5:   # 변수 n 값이 5보다 작거나 같으면 반복
    print(n)
    n += 1
````

````python
# 5 ~ 1 까지 내림차순으로 출력
n = 5
while n >= 1:
    print(n)
    n -= 1
````

````python
# 2 4 6 8 10 출력
n = 1
while n <= 5:
    print(2 * n)
    n += 1
````

````python
n = 2
while n <= 10:
    if n % 2 == 0:
        print(n)
    n += 1
````

````python
# 9 7 5 3 1 출력
n = 5
while n >=1:
    print((2 * n) - 1)
    n -= 1
````

````python
n = 9
while n > 0:
    if n % 2:       #if  n % 2 ==1: 과 같다. 그이유는 1은 True니까.
        print(n)
    n -= 1
````

# `break`와 `continue`

*   `break`: 반복(iteration)을 종료. break가 포함된 가장 가까운 반복문을 빠져나감.
*   `continue`: 다음 반복(iteration)을 수행. continue 아래의 문장들은 건너뛰고, 그 다음 반복과정을 실행.

````python
n = 1
while n <= 5:
    if n == 3:
        break
    print(n)
    n += 1
````

````python
n = 0
while n < 5:
    n += 1
    if n == 3:
        continue
    print(n)
````

# 연습문제

## 1.구구단 2단을 출력

````python
n = 1
while n < 10:
    print(f'2 X {n} = {2 * n}')
    n += 1
````

2. 구구단 2단부터 9단까지 출력

````python
n = 1
m = 2
print('구구단 시작!')
while m <= 9:
    print(f'---- {m}단 ----')
    while n < 10:
        print(f'{m} X {n} = {m * n}')
        n += 1
    m += 1
    n = 1
print('--- 구구단 종료 ---')
````

## *가 하나씩늘어나서 1~5까지 늘어나는 코드

````python
n = 1
while n <= 5:
    print('*' * n)
    n += 1
````

## 반대로 *가 5에서 1까지 줄어드는 코드

````python
n = 5
while n >= 1:
    print('*' * n)
    n -= 1
````

## 구구단 2단은 2 * 2까지.. 9단은 9*9까지 출력코드작성

````python
x = 2
while x < 10:

    y=1
    while y < 10:
        print(f'{x} X {y} = {x * y}')
        if x == y:
            break
        y +=1
    x += 1
````

````python
x = 2
while x < 10:
    print(f'=== {x}단 ===')

    y = 1
    while y <10:
        print(f'{x} X {y} = {x * y}')
        if x == y:
            break
        y += 1
    x += 1
````

````python
x = 2
while x < 10:
    print(f'=== {x}단 ===')

    y = 1
    while y <= x:
        print(f'{x} x {y} = {x * y}')
        y += 1

    print('----------')
    x += 1
````
