```python
#테이블에는 저장된 데이터는 몇년도부터 몇년도까지 조사한 내용일까요?
gapminder.year.unique()
```

```python
#기대 수명이 최댓값인 레코드(row)를 찾으세요.
gapminder[gapminder.lifeExp == max(gapminder.lifeExp)]
```

```python
#인구가 최댓값인 레코드(row)를 찾으세요.
gapminder[gapminder['pop'] == max(gapminder['pop'])]
```

```python
#1인당 GDP가 최댓값인 레코드(row)를 찾으세요.
gapminder[gapminder.gdpPercap == max(gapminder.gdpPercap)]
```

```python
#우리나라의 통계 자료만 출력하세요.
gapminder[gapminder.country == 'Korea, Dem. Rep.']
```

```python
#연도별 1인당 GDP의 최댓값인 레코드를 찾으세요.
gy = gapminder.year.unique()
```

```python
result = []

for y in gy:
    data = gapminder[gapminder.year == y]

    r = data[data.gdpPercap == data.gdpPercap.max()]

    result.append(r)

gdp_max_by_year = pd.concat(result)
```

```python
gdp_max_by_year
```

```python
# 대륙별 1인당 GDP의 최댓값인 레코드를 찾으세요.
c = gapminder.continent.unique()
c
```

```python
result = []
for i in c:
    cdata = gapminder[gapminder.continent == i]

    cmg = cdata[cdata.gdpPercap == max(cdata.gdpPercap)]
    result.append(cmg)

cmg2 = pd.concat(result)
```

```python
cmg2
```

```python
#연도별, 대륙별 인구수를 출력하세요.
gapminder.groupby(['year', 'continent'])['pop'].sum()
```

```python

year_pop = gapminder.groupby('year')['pop'].sum()


max_year = year_pop[year_pop == year_pop.max()]


print(f"인구 가장 많았던 연도: {max_year}")
```

```python
#연도별, 대륙별 기대 수명의 평균을 출력하세요.
gapminder.groupby(by = ['year', 'continent']).lifeExp.mean()
```

```python
#기대 수명이 가장 긴 연도와 대륙은 어디인가요?
ycl = gapminder.groupby(by = ['year', 'continent']).lifeExp.value_counts()
ycl
```

```python
ycl[ycl.lifeExp == ycl.lifeExp.max()]
```

```python
#연도별, 대륙별 1인당 GDP의 평균을 출력하세요.

gapminder.groupby(by = ['year', 'continent']).gdpPercap.mean()
```

```python
#1인당 GDP의 평균이 가장 큰 연도와 대륙은 어디인가요?
g_group = gapminder.groupby(['year', 'continent'])['gdpPercap'].mean()
result = g_group[g_group == g_group.max()]
print(result)
```
