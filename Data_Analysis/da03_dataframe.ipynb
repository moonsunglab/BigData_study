# titanic 데이터셋

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.tail()
```

titanic 데이터프레임 컬럼
*   survived: 생존 여부(0 = 사망, 1 = 생존)
    *   alive 컬럼과 같은 내용(no = 사망, yes = 생존)
*   pclass: 객실 등급
    *   class 컬럼과 같은 내용(1 = First, 2 = Second, 3 = Third,...)
*   sex: 성별(male/female)
*   who: 성인남성(man), 성인여성(woman), 어린이(child).
*   adult_male: 성인 남성 여부(True/False)
*   age: 나이
*   sibsp: 동승한 형제자매 또는 배우자의 수
*   parch: 동승한 부모 또는 자녀의 수
*   alone: 혼자 탑승했는지 여부(True/False)
*   fare: 탑승권 가격. 운임.

```python
woman_survived = titanic[(titanic.who == 'woman') & (titanic.survived == 1)].shape[0] / titanic[titanic.survived == 1].shape[0]
woman_survived
```

```python
child_survived =titanic[(titanic.who == 'child') & (titanic.survived == 1)].shape[0] / titanic[titanic.survived == 1].shape[0]
child_survived
```

### 오쌤 방법

```python
alive_df = titanic[titanic.alive == 'yes']          # 생존자들만 저장된 DF
alive_df
```

```python
alive_df.who.value_counts()
```

```python
titanic.who.value_counts()
```

```python
alive_df.who.value_counts() / titanic.who.value_counts()                        # Series의 특성을 사용해서 정말 쉽게 할 수 있다
```

## 객실 등급별 생존 비율

```python
titanic.pclass.value_counts()
```

```python
alive_df.pclass.value_counts()
```

```python
alive_df.pclass.value_counts() / titanic.pclass.value_counts()
```

## 각 객실 등급(First, Second, Third)에서 남성/여성/어린이 생존 비율

```python
titanic['class'].value_counts()
```

```python
first_class = titanic[titanic['class'] == 'First']
first_class
```

```python
first_class.who.value_counts()
```

```python
first_class_alive = first_class[first_class['survived'] ==1]
first_class_alive.who.value_counts()
```

```python
first_class_alive.who.value_counts()/first_class.who.value_counts()
```

```python
#Second 석 생존율
second_class = titanic[titanic['class'] == 'Second']
```

```python
second_class.who.value_counts()
```

```python
second_class_alive = second_class[second_class['survived'] ==1]
second_class_alive.who.value_counts()
```

```python
second_class_alive.who.value_counts() / second_class.who.value_counts()
```

```python
# Third 생존율
third_class = titanic[titanic['class'] == 'Third']
```

```python
third_class.who.value_counts()
```

```python
third_class_alive = third_class[third_class['survived'] ==1]
third_class_alive.who.value_counts()
```

```python
third_class_alive.who.value_counts() / third_class.who.value_counts()
```

```python
# 세번째 방법
# for c in range(1,4):           # unique == sql에서의 dinstinct
for c in sorted(titanic.pclass.unique()):
    print(f'--- class {c} ---')
    subset = titanic[titanic.pclass == c]               # 객실 등급의 승객
    alive_subset = subset[subset.alive == 'yes']        # 객실에서 생존자
    print(alive_subset.who.value_counts() / subset.who.value_counts())
```
